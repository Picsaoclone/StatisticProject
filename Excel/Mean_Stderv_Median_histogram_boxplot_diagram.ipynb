```python
import numpy as np
import statistics as st
import pandas as pd 
```

```python
# Đọc từng file CSV với mã hóa 'ISO-8859-1' hoặc 'cp1252'
df_TCB = pd.read_csv('financial_data_TCB.csv',encoding='ISO-8859-1')

# Kiểm tra dữ liệu của từng DataFrame (nếu cần)
print("Data TCB:")
print(df_TCB.head())
```

```python
import matplotlib.pyplot as plt

# Danh sách các cột muốn vẽ histogram
columns_to_plot = [
    'roe', 'roa', 'bookValuePerShare', 'interestMargin', 'badDebtPercentage', 
    'provisionOnBadDebt', 'equityOnTotalAsset', 'equityOnLoan', 'assetOnEquity', 
    'postTaxOnToi', 'loanOnEarnAsset', 'loanOnAsset', 'badDebtOnAsset', 
    'payableOnEquity', 'cancelDebt', 'bookValuePerShareChange', 'creditGrowth'
]

# Vẽ histogram cho từng cột
for col in columns_to_plot:
    if col in df_TCB.columns:
        plt.figure(figsize=(8, 5))
        plt.hist(df_TCB[col].dropna(), bins=20, edgecolor='black')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.title(f'Histogram of {col}')
        plt.show()
    else:
        print(f"Column '{col}' not found in the DataFrame.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Vẽ boxplot ở trên
sns.boxplot(data=df_TCB, x='roe', ax=ax1)  # Thay 'roe' bằng tên cột bạn muốn vẽ
ax1.set_xlabel('ROE')
ax1.set_title('Boxplot of ROE')

# Vẽ histogram ở dưới
ax2.hist(df_TCB['roe'].dropna(), bins=20, edgecolor='black')  # Thay 'roe' bằng tên cột bạn muốn vẽ
ax2.set_xlabel('ROE')
ax2.set_ylabel('Frequency')
ax2.set_title('Histogram of ROE')

plt.tight_layout()
plt.show()
```
