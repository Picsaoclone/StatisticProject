```python
import numpy as np
import statistics as st
import pandas as pd 
```

```python
# Đọc từng file CSV với mã hóa 'ISO-8859-1' hoặc 'cp1252'
df_VCB = pd.read_csv('financial_data_VCB_filtered.csv',encoding='ISO-8859-1')

# Kiểm tra dữ liệu của từng DataFrame (nếu cần)
print("Data VCB:")
print(df_VCB.head())
```

```python
mean_roe = np.mean(df_VCB['roe'])
print(f"roe mean: {mean_roe}")
print(f"roe standard deviation: {st.stdev(df_VCB['roe'])}")
print(f"roe median: {st.median(df_VCB['roe'])}")

mean_roa = np.mean(df_VCB['roa'])
print(f"roa mean: {mean_roa}")
print(f"roa standard deviation: {st.stdev(df_VCB['roa'])}")
print(f"roa median: {st.median(df_VCB['roa'])}")

mean_bookValuePerShare = np.mean(df_VCB['bookValuePerShare'])
print(f"bookValuePerShare mean: {mean_bookValuePerShare}")
print(f"bookValuePerShare standard deviation: {st.stdev(df_VCB['bookValuePerShare'])}")
print(f"bookValuePerShare median: {st.median(df_VCB['bookValuePerShare'])}")

mean_interestMargin = np.mean(df_VCB['interestMargin'])
print(f"interestMargin mean: {mean_interestMargin}")
print(f"interestMargin standard deviation: {st.stdev(df_VCB['interestMargin'])}")
print(f"interestMargin median: {st.median(df_VCB['interestMargin'])}")

mean_badDebtPercentage = np.mean(df_VCB['badDebtPercentage'])
print(f"badDebtPercentage mean: {mean_badDebtPercentage}")
print(f"badDebtPercentage standard deviation: {st.stdev(df_VCB['badDebtPercentage'])}")
print(f"badDebtPercentage median: {st.median(df_VCB['badDebtPercentage'])}")

mean_provisionOnBadDebt = np.mean(df_VCB['provisionOnBadDebt'])
print(f"provisionOnBadDebt mean: {mean_provisionOnBadDebt}")
print(f"provisionOnBadDebt standard deviation: {st.stdev(df_VCB['provisionOnBadDebt'])}")
print(f"provisionOnBadDebt median: {st.median(df_VCB['provisionOnBadDebt'])}")

mean_equityOnTotalAsset = np.mean(df_VCB['equityOnTotalAsset'])
print(f"equityOnTotalAsset mean: {mean_equityOnTotalAsset}")
print(f"equityOnTotalAsset standard deviation: {st.stdev(df_VCB['equityOnTotalAsset'])}")
print(f"equityOnTotalAsset median: {st.median(df_VCB['equityOnTotalAsset'])}")


```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Vẽ boxplot ở trên
sns.boxplot(data=df_VCB, x='roe', ax=ax1)  
ax1.set_xlabel('ROE')
ax1.set_title('Boxplot of ROE')

# Vẽ histogram ở dưới
ax2.hist(df_VCB['roe'].dropna(), bins=20, edgecolor='black')  
ax2.set_xlabel('ROE')
ax2.set_ylabel('Frequency')
ax2.set_title('Histogram of ROE')

plt.tight_layout()
plt.show()
```
