```python
import numpy as np
import statistics as st
import pandas as pd 
```

```python
# Đọc từng file CSV với mã hóa 'ISO-8859-1' hoặc 'cp1252'
df_TCB = pd.read_csv('financial_data_TCB_filtered.csv',encoding='ISO-8859-1')

# Kiểm tra dữ liệu của từng DataFrame (nếu cần)
print("Data TCB:")
print(df_TCB.head())
```

```python
mean_roe = np.mean(df_TCB['roe'])
print(f"roe mean: {mean_roe}")
print(f"roe standard deviation: {st.stdev(df_TCB['roe'])}")
print(f"roe median: {st.median(df_TCB['roe'])}")

mean_roa = np.mean(df_TCB['roa'])
print(f"roa mean: {mean_roa}")
print(f"roa standard deviation: {st.stdev(df_TCB['roa'])}")
print(f"roa median: {st.median(df_TCB['roa'])}")

mean_bookValuePerShare = np.mean(df_TCB['bookValuePerShare'])
print(f"bookValuePerShare mean: {mean_bookValuePerShare}")
print(f"bookValuePerShare standard deviation: {st.stdev(df_TCB['bookValuePerShare'])}")
print(f"bookValuePerShare median: {st.median(df_TCB['bookValuePerShare'])}")

mean_interestMargin = np.mean(df_TCB['interestMargin'])
print(f"interestMargin mean: {mean_interestMargin}")
print(f"interestMargin standard deviation: {st.stdev(df_TCB['interestMargin'])}")
print(f"interestMargin median: {st.median(df_TCB['interestMargin'])}")

mean_badDebtPercentage = np.mean(df_TCB['badDebtPercentage'])
print(f"badDebtPercentage mean: {mean_badDebtPercentage}")
print(f"badDebtPercentage standard deviation: {st.stdev(df_TCB['badDebtPercentage'])}")
print(f"badDebtPercentage median: {st.median(df_TCB['badDebtPercentage'])}")

mean_provisionOnBadDebt = np.mean(df_TCB['provisionOnBadDebt'])
print(f"provisionOnBadDebt mean: {mean_provisionOnBadDebt}")
print(f"provisionOnBadDebt standard deviation: {st.stdev(df_TCB['provisionOnBadDebt'])}")
print(f"provisionOnBadDebt median: {st.median(df_TCB['provisionOnBadDebt'])}")

mean_equityOnTotalAsset = np.mean(df_TCB['equityOnTotalAsset'])
print(f"equityOnTotalAsset mean: {mean_equityOnTotalAsset}")
print(f"equityOnTotalAsset standard deviation: {st.stdev(df_TCB['equityOnTotalAsset'])}")
print(f"equityOnTotalAsset median: {st.median(df_TCB['equityOnTotalAsset'])}")


```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Vẽ boxplot ở trên
sns.boxplot(data=df_TCB, x='roe', ax=ax1)  
ax1.set_xlabel('ROE')
ax1.set_title('Boxplot of ROE')

# Vẽ histogram ở dưới
ax2.hist(df_TCB['roe'].dropna(), bins=20, edgecolor='black')  
ax2.set_xlabel('ROE')
ax2.set_ylabel('Frequency')
ax2.set_title('Histogram of ROE')

plt.tight_layout()
plt.show()
```
